For many pages


```
base_url = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc='
loc = 'Newport+Beach ,+CA'
page = 10
url=base_url + loc + "&start=" + str(page)
```

```
current_page = 0
```

```
while current_page < 201:
  print(current_page)
  url = base_url + loc + "&start=" + str(current_page)
  yelp_r = requests.get(url)
  yelp_soup = BeautifulSoup(yelp_r.text, 'html.parser')
  businesses = yelp_soup.findAll('div',{'class':'businessName__09f24__3Ml2X display--inline-block__09f24__3SvIn border-color--default__09f24__3Epto'})
  
  file_path = 'yelp-{loc}-2.txt'.format(loc=loc)
  
  with open(file_path,"a") as textfile:
      businesses = yelp_soup.findAll('div',{'class':' businessName__09f24__3Ml2X display--inline-block__09f24__3SvIn border-color--default__09f24__3Epto'})
      for biz in businesses:
          #print(biz)
          title = biz.findAll('a',{'class':'css-166la90'})[0].text
          print(title)
          page_line = "{title}\n".format(title=title)
          #page_line = "{title}\n{address}\n{phone}".format(title=title,address=address,phone=phone )
          textfile.write(page_line)
  current_page += 10
```
